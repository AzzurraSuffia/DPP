## Errors and failures

```python
social_anonymizer = SocialAnonymizer()
```

```python
fails = []
errors = []
for seed in range(100):
    G_random = nx.gnp_random_graph(n=15, p=0.08, seed= seed)
    #plot_component(G_random, "original graph")
    try:
        G_anon, eq_dict = social_anonymizer.anonymize_graph(G_random, k=4, alpha=0, beta=1, gamma=1)
        #plot_component(G_anon, "anon graph")
        if check_isomorphic_classes(G_anon, eq_dict) != []:
            fails.append(seed)
    except Exception as e:
        errors.append(seed)

print("Seeds that failed: ", fails)
print("Seeds that returned an error: ", errors)
```

```python
n = 20
m = 1
count = 0
error = 0
fails = []
errors = []
for seed in range(50):
    G_ba = nx.barabasi_albert_graph(n, m, seed=seed)
    social_anonymizer = SocialAnonymizer()
    try:
        G_anon, eq_dict = social_anonymizer.anonymize_graph(G_ba, k=3, alpha=0, beta=1, gamma=1)
        if len(check_isomorphic_classes(G_anon, eq_dict)) == 0:
            count += 1
        else:
            fails.append(seed)
    except Exception:
        error += 1
        errors.append(seed)

print(f"Success rate: {count}/{50}")
print(f"Error rate: {error}")
print("Seeds that failed: ", fails)
print("Seeds that returned an error: ", errors)
```
